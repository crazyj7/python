```python
import itertools

def solution(number, k):
    answer = ''
    comb = list(itertools.combinations(range(len(number)), k))
    numlist=set()
    for ri in comb:
        number2 = list(number)
        for i in ri:
            number2[i]=''
#         numlist.add(int(''.join(number2)))
        numlist.add(''.join(number2))
#     print(numlist)
    answer = max(numlist)
    return answer
```

```python
import numpy as np
def solution(number, k):
    answer = ''
    if k==0:
        return number
    numlen = len(number)
    number = np.asarray(list(number))
    i=0
    
    # perf: 9가 아닌 것을 미리 제거하자.
    not9 = []
    i9=0
    tmp = []
    while i9<(numlen-k):
        if number[i9]=='9':
            # commit
            if len(not9)+len(tmp)>=k:
                break
            not9.extend(tmp)
            i9+=1
            tmp=[]
        else:
            tmp.append(i9)
            i9+=1
    print('not9 count=', len(not9))
    if len(not9)>0:
        number = np.delete(number, not9)
        k-=len(not9)
    
    #    
    while i<(numlen-k):
        # print(number)
        if number[i]=='9':
            i+=1
            continue
        bmaxfound=False
        maxi=0
        for m in range(i, i+k+1):
            if number[m]=='9':
                maxi = m-i
                bmaxfound=True
                break
        if not bmaxfound:
            maxi = np.argmax(number[i:i+k+1])
        if maxi>0:
            number = np.delete(number, range(i, i+maxi) )
            k-=maxi
            numlen-=maxi
        i+=1
        if k==0:
            break
    # print('remaink=', k)
    if k>0:
        number=number[:-k]
    # print(number)    
    answer = ''.join(number)
    return answer
```
