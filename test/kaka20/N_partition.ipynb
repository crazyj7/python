```python
def findnth(string, substring, n):
    parts = string.split(substring, n + 1)
    if len(parts) <= n + 1:
        return -1
    return len(string) - len(parts[-1]) - len(substring)
```

```python
import itertools

minpart=1 # 최소 숫자개수=1
maxpart=8 # 최대숫자개수=8
idebug=0

def solution(N, number):
    answer = 0
    bfound = False
    N = str(N)

    # i=숫자 개수 (1~9)
    for i in range(minpart,(maxpart+1)):
        pools=[ x for x in range(i)] # 숫자 개수만큼 넘버링. (0~9까지만 가능.)
        print('')
        print('pools=', pools)
        for j in range(i):   # 0~(i-1)까지 
            p = itertools.combinations(pools[:-1], j)  # 마지막 넘버링을 빼고, 조합. 
            # 즉, 숫자개수-1 에서의 조합. 숫자3개면 2C0, 2C1, 2C2.  : '', 0, 1, 01.  이 위치로 partition을 한다.  
            strn = [N,' ']*i  # 숫자 반복 스트링. (공백추가.)
#             print('')
#             print('input=', ''.join(strn))
            for pi in p:
                strn = [N,' ']*i
                for el in pi:
                    strn[el*2+1]='|'
                equation = ''.join(strn).replace(' ', '') # operator = |
#                 print('equation1=', equation)

                # operator change...
                # 4^ (number of operators)
                terms = equation.split('|')
                n_terms = len(terms)
                opers = list(itertools.product('+-*/', repeat=n_terms-1))
                # 연산자를 텀 사이에 위치시키기 위한 모든 경우의 수. 프로덕트. 반복. (반복 순열)
                # ++++, +++-, +++*, ...

                for o in opers:
                    # operators list. one case.
                    # ex) +, +, +, +
                    eq = []
                    for it, t in enumerate(terms):
                        #  55 + 5 - 555 + 5 * 55 
                        eq.append(' ')
                        eq.append(t)
                        eq.append(' ')
                        if it<n_terms-1:
                            eq.append(o[it])
                    equationr = ''.join(eq)
                    if idebug>0:
                        print('debug =[%s]'%equationr)
                        
                    # parenthesis
                    # 괄호 1개...?개
                    if n_terms>2: # 텀이 3개는 되어야 괄호 사용가능.
                        # 1p, 1o ; 괄호 1개, operator 1개.
                        # pp = p posotion.
#                         print('opers=', o)
                        for pp in range(len(o)): # ++이면 0, 1
#                             print('pp=', pp)
                            equationp = equationr
                            ps = findnth(equationp, ' ', pp*2)
                            pe = findnth(equationp, ' ', pp*2+3)
#                             print(pp, ps, pe)
                            equationp = equationp[:ps]+'('+equationp[ps+1:]
                            if len(equationp)>pe+1:
                                equationp = equationp[:pe]+')'+equationp[pe+1:]
                            else:
                                equationp = equationp[:pe]+')'
                            
                            equation2 = equationp.replace('/', '//')
                            try:
                                result = eval(equation2)
                            except:
#                                 print('error!')
                                result = 0;
                            if idebug>0:
                                print('equation2*=', equation2, '=',result)
                            if result==number:
                                bfound=True
                                break
                            # (5 + 5)+ 5 + 5 + 5 + 5 
                            
                                
                    # end of if i>2 
                    else:
                        equation2 = equationr.replace('/', '//')
                        try:
                            result = eval(equation2)
                        except:
#                             print('error!')
                            result = 0;
                        if idebug>0:
                            print('equation2=', equation2, '=',result)
                        if result==number:
                            bfound=True
                            break
                if bfound:
                    break
            if bfound:
                break
        if bfound:
            break
    result = eval(equation2)
    print('equation2=', equation2, '=',result)
    print('bfound=', bfound)
    answer=i
    print('answer=', answer)
    return answer
```

```python
import itertools
from collections import Counter
from datetime import datetime


minpart=1 # 최소 숫자개수=1
maxpart=7 # 최대숫자개수=8
idebug=0

def solution(N, number):
    answer = 0
    bfound = False
    N = str(N)
    t_start = datetime.now().timestamp()
 

    # i=숫자 개수 (1~9)
    for i in range(minpart,(maxpart+1)):
        pools=[ x for x in range(i)] # 숫자 개수만큼 넘버링. (0~9까지만 가능.)
        print('digit count=', i)
        if idebug:
            print('')
            print('pools=', pools)
        for j in range(i):   # 0~(i-1)까지 
            p = itertools.combinations(pools[:-1], j)  # 마지막 넘버링을 빼고, 조합. 
            # 즉, 숫자개수-1 에서의 조합. 숫자3개면 2C0, 2C1, 2C2.  : '', 0, 1, 01.  이 위치로 partition을 한다.  
            strn = [N,' ']*i  # 숫자 반복 스트링. (공백추가.)
#             print('')
#             print('input=', ''.join(strn))
            for pi in p:
                strn = [N,' ']*i
                for el in pi:
                    strn[el*2+1]='|'
                equation = ''.join(strn).replace(' ', '') # operator = |
#                 print('equation1=', equation)

                # operator change...
                # 4^ (number of operators)
                terms = equation.split('|')
                terms = list( map(int, terms))
                n_terms = len(terms)
                n_opers = n_terms-1
                opers = list(itertools.product('+-*/', repeat=n_opers))
                o=''
                r=''
                # 연산자를 텀 사이에 위치시키기 위한 모든 경우의 수. 프로덕트. 반복. (반복 순열)
                # ++++, +++-, +++*, ...
                if idebug:
                    print('terms=', terms)
                    print('n_terms=', n_terms)
                    
                for o in opers:
                    o = list(o)
                    if idebug:
                        print('operator=', o)
                    # 동급 op들로만 구성시 order는 의미없다.
                    bsamelevel=False
                    # +만 존재하면 우선순위가 의미없다.
                    # *만 존재해도 우선순위 무의미.
                    # -, /는 다르다. a-(b-c) (a-b)-c
#                     if '*' not in o and '/' not in o:
#                         bsamelevel=True
#                     elif '+' not in o and '-' not in o:
#                         bsamelevel=True
                    ocount = Counter(o)
                    ocountkeys = list(ocount.keys())
                    if len(ocountkeys)==1 :
                        if ocountkeys[0]=='+' or ocountkeys[0]=='*':
                            bsamelevel=True
                    
                    # make orders
                    if bsamelevel:
                        orders = [ list(range(n_opers)) ]
                    else:
                        orders = list(itertools.permutations(range(n_opers), len(o)))
                    if idebug:
                        print('orders=', orders)
                    for r in orders:
                        r = list(r)
                        if idebug:
                            print('order=', r)
                        terms2 = terms[:]
                        opers2 = o[:]
                        # calculate the equation.
                        while len(opers2)>0:
                            if len(r)==0:
                                ret=terms2[0]
                            index = r.index(0)
                            t1 = int(terms2[index])
                            t2 = int(terms2[index+1])
                            op = opers2[index]
                            if idebug:
                                print('calc:', t1, op, t2)
                            ret = 0 
                            
                            if op=='+':
                                ret=t1+t2
                            elif op=='-':
                                ret = t1-t2 
                            elif op=='*':
                                ret = t1*t2
                            elif op=='/':
                                if t2==0:
                                    if idebug:
                                        print('error.divz')
                                    terms2[0]=0
                                    ret=0
                                    break
                                else:
                                    ret=t1//t2
                            terms2[index]=ret
                            del terms2[index+1]
                            del r[index]
                            del opers2[index]
                            r2=[]
                            for x in r:
                                r2.append(x-1)
                            r=r2
                        ret = terms2[0]
                        if idebug:
                            print('result=', ret)
                        if ret==number:
                            bfound=True
                            break
                    if bfound:
                        break
                
                if ret==number:
                    bfound=True
                    if idebug:
                        print('\nsolution')
                        print('terms=', terms)
                        print('n_terms=', n_terms)
                        print('operator=', o)
                        print('order=', r)
                    break
                if bfound:
                    break
            if bfound:
                break
        if bfound:
            break
    print('\nbfound=', bfound)
    if not bfound:
        answer=-1
    else:
        answer=i
    print('answer=', answer)
    t_end = datetime.now().timestamp()
    print('time=', t_end-t_start)
    return answer
```
